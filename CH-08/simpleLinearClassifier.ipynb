```python
import cv2
import numpy as np
```

```python
labels = ['dog', 'cat', 'panda']
temp = np.random.seed(1)
```

```python
w = np.random.randn(3, 3072)
b = np.random.randn(3)
```

```python
original_image = cv2.imread('beagle.png')
```

```python
image = cv2.resize(original_image, (32, 32)).flatten()
```

```python
scores = w.dot(image) + b
```

```python
image.shape
```

```python
scores
```

```python
for label, score in zip(labels, scores):
    print('INFO: {}: {:.2f}'.format(label, score))
```

```python
cv2.putText(original_image, 'Label: {}'.format(labels[np.argmax(scores)]),(10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
```
